```python
from collections import Counter

# Sample list of names
names = ["Alice", "Bob", "Alice", "Charlie", "Bob", "David", "Alice", "Bob"]

# Use Counter to count name occurrences
name_counts = Counter(names)

# Print the count for each unique name
for name, count in name_counts.items():
    print(f"{name}: {count} times")
```

```python
# Sample lists of names
list1 = ["Alice", "Bob", "Alice", "Charlie", "Bob", "David", "Alice", "Bob"]
list2 = ["Alice", "Eve", "Frank", "Alice", "Charlie", "Eve"]

# Combine the lists into one
all_lists = list1 + list2

# Create a dictionary to count name occurrences
name_counts = {}

# Count name occurrences in all_lists
for name in all_lists:
    if name in name_counts:
        name_counts[name] += 1
    else:
        name_counts[name] = 1

# Specify the name you want to check
target_name = "Alice"

# Get the count for the specified name
if target_name in name_counts:
    total_count = name_counts[target_name]
    print(f"{target_name} has been repeated {total_count} times in all lists.")
else:
    print(f"{target_name} has not been repeated in any list.")
```

```python
def detect_outliers(data):
    outliers = []
    # outliers zitakua zile ziko beyond two standard deviations
    threshold = 2
    mean = np.mean(data)
    # print(f"Mean of the {mwezi} is {mean}")
    std = np.std(data)
    # print(f"Standard deviation ya {mwezi} {std}")
    for i in data:
        z_score = (i - mean) / std
        # print(f"Z_score for {i} is {z_score}")
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```
